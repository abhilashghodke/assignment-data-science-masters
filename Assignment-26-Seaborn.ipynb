Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans--


Seaborn is a popular Python data visualization library that is built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five types of plots you can create using Seaborn, along with their uses:

1. Scatter Plot:

- Use: Scatter plots are used to display the relationship between two continuous variables. Each point on the plot represents a data point, making it easy to identify patterns or trends.
- Function: seaborn.scatterplot()

2. Line Plot:

- Use: Line plots are used to visualize the trend of a continuous variable over a continuous interval or time. They are often used to show data changes over time or any other continuous sequence.
- Function: seaborn.lineplot()

3. Bar Plot:

- Use: Bar plots are used to compare categorical data. They can be used to show the distribution of data across different categories or to compare values of a single category across different groups.
- Function: seaborn.barplot()

4. Histogram:

- Use: Histograms are used to represent the distribution of a single variable. They group data into intervals (bins) and show the frequency or count of data points falling into each bin.
- Function: seaborn.histplot()

5. Box Plot (Box-and-Whisker Plot):

- Use: Box plots are used to display the distribution of data and identify potential outliers. They provide information about the median, quartiles, and possible skewness in the data.
- Function: seaborn.boxplot()

These are just a few examples of the types of plots you can create using Seaborn. The library offers many more types of plots and customization options for making your visualizations more informative and appealing.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Ans--

Certainly! To create a line plot using the "fmri" dataset from Seaborn with "timepoint" on the x-axis and "signal" on the y-axis for different events and regions, you can follow these steps:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Add title and labels
plt.title("FMRI Signal over Time for Different Events and Regions")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the legend
plt.legend(title="Event")

# Show the plot
plt.show()
```

In this code, we use the sns.load_dataset("fmri") function to load the "fmri" dataset from Seaborn. We then create a line plot using sns.lineplot() where we set the "timepoint" column as the x-axis data, the "signal" column as the y-axis data, and use the "event" column for coloring lines and the "region" column for line styles. The hue parameter is used to distinguish different events with different colors, and the style parameter is used to differentiate regions with different line styles. Finally, we add a title, labels, and legend to the plot for better readability.

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

Ans--


Here's how you can load the "titanic" dataset from Seaborn and plot two box plots using the specified columns:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Create a figure with subplots
plt.figure(figsize=(12, 5))

# Create the first box plot for 'age' vs. 'pclass'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Create the second box plot for 'fare' vs. 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

In this code, we use the sns.load_dataset("titanic") function to load the "titanic" dataset from Seaborn. We then create a figure with two subplots side by side. For each subplot, we use sns.boxplot() to create a box plot. The first subplot shows the distribution of "age" for different passenger classes ("pclass"), and the second subplot shows the distribution of "fare" for different passenger classes.

We set titles, x-axis labels, and y-axis labels for each subplot. Finally, plt.tight_layout() ensures that the plots are neatly arranged, and plt.show() displays the plots.

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

Ans--

 Here's how you can use the "diamonds" dataset from Seaborn to plot a histogram for the 'price' column, using the hue parameter for the 'cut' column:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram with hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')

# Add title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")

# Show the legend
plt.legend(title="Cut")

# Show the plot
plt.show()
```

In this code, we use the sns.load_dataset("diamonds") function to load the "diamonds" dataset from Seaborn. We then create a histogram using sns.histplot(), where we set the 'price' column as the x-axis data and the 'cut' column as the hue, which will differentiate the histogram bars by the diamond cut. The multiple='stack' parameter ensures that the histograms are stacked based on the 'cut' values.

We add a title, x-axis label, y-axis label, and a legend to the plot for better readability. Finally, plt.show() displays the plot.

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

Here's how you can use the "iris" dataset from Seaborn to plot a pair plot, using the hue parameter for the "species" column:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a pair plot with hue
sns.pairplot(data=iris, hue='species')

# Add title
plt.title("Pair Plot of Iris Dataset by Species")

# Show the plot
plt.show()
```

In this code, we use the sns.load_dataset("iris") function to load the "iris" dataset from Seaborn. We then create a pair plot using sns.pairplot(), where we set the hue parameter to the 'species' column. This will color the scatter plots and histograms based on the different species of iris flowers.

We add a title to the plot and then use plt.show() to display the plot.

The pair plot will show scatter plots for each pair of numerical columns in the dataset, along with histograms for each individual column. The scatter plots will be differentiated by colors according to the different species of iris flowers.

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

Ans--

Here's how you can use the "flights" dataset from Seaborn to plot a heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data=flights_pivot, annot=True, fmt='d', cmap='YlGnBu')

# Add title and labels
plt.title("Passenger Traffic Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```
