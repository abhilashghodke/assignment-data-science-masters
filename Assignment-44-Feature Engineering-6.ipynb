Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

Ans--

To calculate the Pearson correlation coefficient between two variables and interpret the result using a Seaborn dataset, you can follow these steps. In this example, I'll use the built-in "tips" dataset from Seaborn, which contains data on restaurant tips and total bills. You can replace it with your dataset as needed:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Seaborn dataset (replace with your dataset)
data = sns.load_dataset("tips")

# Calculate the Pearson correlation coefficient
corr_coefficient = data["total_bill"].corr(data["tip"])

# Create a scatter plot
sns.scatterplot(x="total_bill", y="tip", data=data)
plt.title(f"Pearson Correlation Coefficient: {corr_coefficient:.2f}")

# Interpret the result
if corr_coefficient > 0:
    interpretation = "There is a positive linear relationship between total bill and tip amount."
elif corr_coefficient < 0:
    interpretation = "There is a negative linear relationship between total bill and tip amount."
else:
    interpretation = "There is no linear relationship between total bill and tip amount."

print(interpretation)
plt.show()
```

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

Ans--

To calculate Spearman's rank correlation between two variables using a Seaborn dataset, you can follow these steps. In this example, I'll use a hypothetical dataset since Seaborn doesn't have a built-in dataset for this specific scenario. You can replace it with your dataset as needed:

```python
import seaborn as sns
import pandas as pd
import scipy.stats as stats

# Create a hypothetical dataset (replace with your data)
data = {
    "Sleep Hours": [7, 6, 8, 5, 6, 7, 9, 8, 5, 10],
    "Job Satisfaction": [8, 7, 9, 5, 6, 8, 9, 7, 6, 10],
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Calculate Spearman's rank correlation
spearman_corr, _ = stats.spearmanr(df["Sleep Hours"], df["Job Satisfaction"])

# Interpret the result
if spearman_corr > 0:
    interpretation = "There is a positive monotonic relationship between sleep hours and job satisfaction."
elif spearman_corr < 0:
    interpretation = "There is a negative monotonic relationship between sleep hours and job satisfaction."
else:
    interpretation = "There is no monotonic relationship between sleep hours and job satisfaction."

print(f"Spearman's Rank Correlation: {spearman_corr:.2f}")
print(interpretation)
```

In this code, we create a hypothetical dataset with two columns: "Sleep Hours" and "Job Satisfaction." We then calculate Spearman's rank correlation between these two variables using the spearmanr function from the scipy.stats library. The interpretation will help you understand the direction of the monotonic relationship based on the sign of the correlation coefficient.

Replace the hypothetical dataset with your actual data when working with your specific dataset. Spearman's rank correlation assesses monotonic (non-linear) relationships, which may be more suitable for variables that do not have a linear relationship.

Q6. A company is interested in examining the relationship between the number of sales calls made per day
and the number of sales made per week. The company collected data on both variables from a sample

Ans

To examine the relationship between the number of sales calls made per day and the number of sales made per week using Python, you can follow these steps:

1. **Gather and Organize Data**: First, you need to gather and organize the data into a suitable format. Create a dataset with two columns: one for the number of sales calls made per day and another for the number of sales made per week.

2. **Load Data**: Use a data analysis library like Pandas to load the data into a DataFrame.

3. **Data Exploration**: Calculate summary statistics and visualize the data to get a sense of its distribution and any potential outliers.

4. **Calculate Pearson Correlation**: Calculate the Pearson correlation coefficient to quantify the strength and direction of the linear relationship between the two variables.

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Sample data (replace with your own dataset)
data = {
    'Sales_Calls_Per_Day': [10, 15, 20, 8, 25, 12, 18, 30, 22, 7],
    'Sales_Per_Week': [50, 70, 80, 40, 90, 60, 75, 100, 85, 35]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Data Exploration
print("Summary Statistics:")
print(df.describe())

# Scatterplot
plt.scatter(df['Sales_Calls_Per_Day'], df['Sales_Per_Week'])
plt.xlabel('Sales Calls Per Day')
plt.ylabel('Sales Per Week')
plt.title('Relationship Between Sales Calls and Sales')
plt.show()

# Calculate Pearson correlation coefficient
corr_coefficient, _ = pearsonr(df['Sales_Calls_Per_Day'], df['Sales_Per_Week'])
print(f"Pearson Correlation Coefficient: {corr_coefficient:.2f}")
```

Replace the `data` dictionary with your own dataset. This code loads the data, calculates summary statistics, creates a scatterplot to visualize the relationship, and calculates the Pearson correlation coefficient. The correlation coefficient will indicate the strength and direction of the linear relationship between sales calls per day and sales per week.
