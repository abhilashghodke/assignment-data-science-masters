Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans--

Matplotlib is a popular data visualization library in Python. It provides a wide range of functions for creating static, interactive, and animated visualizations in various formats and styles. Matplotlib is commonly used for creating 2D plots, graphs, and charts to visually represent data and insights.

Usage:

- Matplotlib is used to create various types of plots and charts to visualize data, making it easier to understand trends, patterns, and relationships within the data.
- It is widely used in scientific research, data analysis, engineering, finance, and many other fields to communicate data-driven insights effectively.

Five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line Plot: A basic plot used to display data points connected by straight lines. It's commonly used to show trends over time or relationships between variables.
2. Scatter Plot: A plot used to display individual data points as markers. It's helpful for visualizing the distribution and relationship between two variables.
3. Bar Plot: A plot used to represent categorical data with rectangular bars. It's useful for comparing the quantities of different categories.
4. Histogram: A plot used to display the frequency distribution of continuous data. It divides the data into bins and shows the frequency of data points within each bin.
5. Pie Chart: A circular chart used to represent proportions of a whole. It's suitable for displaying the relative sizes of different categories in a dataset.

These are just a few examples of the types of plots that can be created using the Pyplot module of Matplotlib. It provides a wide variety of customization options to create visually appealing and informative visualizations.

```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.
```

Ans--

A scatter plot is a type of data visualization that uses individual data points to represent the values obtained for two different variables. Each data point is represented by a marker on the plot, and its position is determined by the values of the two variables.

Here's how you can generate and plot a scatter plot using the provided code and add a title, xlabel, and ylabel using Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Setting a random seed for reproducibility
np.random.seed(3)

# Generating data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Creating the scatter plot
plt.scatter(x, y, color='blue', marker='o', label='Data Points')

# Adding title and labels
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Displaying legend
plt.legend()

# Displaying the plot
plt.show()
```

In this code, plt.scatter(x, y) creates a scatter plot with the data points specified by arrays x and y. The color parameter specifies the color of the markers, and the marker parameter specifies the marker style.

The plt.title(), plt.xlabel(), and plt.ylabel() functions are used to set the title, x-axis label, and y-axis label, respectively. The plt.legend() function is used to display the legend indicating what the markers represent.

Finally, plt.show() displays the plot.

```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
    
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

Ans--

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It's particularly useful when you want to compare multiple plots side by side.

Here's how you can use the subplot() function to create four line plots using the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2, label='Line 2')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3, label='Line 3')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4, label='Line 4')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Adjusting layout
plt.tight_layout()

# Displaying the plots
plt.show()
```

In this code, we create a 2x2 grid of subplots using the plt.subplot() function. The first argument specifies the number of rows, the second argument specifies the number of columns, and the third argument specifies the index of the current subplot. We then plot each line using the plt.plot() function within the corresponding subplot. The plt.tight_layout() function is used to improve the spacing between subplots. Finally, plt.show() displays the plots.

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

Ans--

A bar plot is a data visualization technique used to represent categorical data using rectangular bars. Each bar's length or height is proportional to the value it represents. Bar plots are commonly used to compare different categories and visualize the relationships between them.

Usage:

- Bar plots are used to compare and display the relative sizes or quantities of different categories.
- They are effective in illustrating patterns, trends, or differences in data across categories.
- Bar plots can be used for both simple and complex datasets to provide insights into categorical data.

Here's how you can create a bar plot and a horizontal bar plot using the provided data using Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating a bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='blue')
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.show()

# Creating a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='green')
plt.title('Company Profits (Horizontal)')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.show()
```

In the code above, plt.bar() is used to create a vertical bar plot, and plt.barh() is used to create a horizontal bar plot. The company array is used as the x-axis labels, and the profit array is used as the heights (vertical bar plot) or widths (horizontal bar plot) of the bars. The color, title, xlabel, and ylabel functions are used to customize the appearance and labels of the plots. Finally, plt.show() displays the plots.

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

Ans--

A box plot, also known as a box-and-whisker plot, is a data visualization method used to display the distribution of a dataset along with its key statistical characteristics. It provides a visual summary of the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum values of a dataset. Additionally, it can show potential outliers in the data.

Usage:

- Box plots are used to quickly identify the central tendency and spread of a dataset.
- They are helpful in comparing distributions and identifying potential outliers.
- Box plots provide insights into the skewness and symmetry of the data.

Here's how you can create a box plot using the provided data using Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot')
plt.xlabel('Box')
plt.ylabel('Value')
plt.show()
```

In the code above, plt.boxplot() is used to create the box plot. The box1 and box2 arrays are passed as arguments to create two boxes side by side. The labels parameter is used to label the boxes. The title, xlabel, and ylabel functions are used to customize the appearance and labels of the plot. Finally, plt.show() displays the box plot.
